```python
# Task 2: Mark the success/failed launches for each site on the map

# Initialize map centered on continental US
spacex_map = folium.Map(location=[30, -100], zoom_start=4)

# Create a MarkerCluster so markers group automatically when zoomed out
marker_cluster = MarkerCluster().add_to(spacex_map)

# Iterate through all launches and add a marker for each
for index, row in spacex_df.iterrows():
    lat, lon = row['Lat'], row['Long']
    outcome = row['class']      # 1 = success, 0 = failure
    site = row['Launch Site']
    
    # Choose color by success/failure
    color = 'green' if outcome == 1 else 'red'
    
    # Popup text with details about the launch
    popup_text = f"""
    <b>Launch Site:</b> {site}<br>
    <b>Outcome:</b> {'Success ✅' if outcome == 1 else 'Failure ❌'}
    """

    # Add marker to map
    folium.CircleMarker(
        location=[lat, lon],
        radius=5,
        color=color,
        fill=True,
        fill_opacity=0.7,
        popup=folium.Popup(popup_text, max_width=300)
    ).add_to(marker_cluster)

# Display the map
spacex_map
```

```python
# Create the base map centered in the US
site_map = folium.Map(location=[30, -100], zoom_start=4)

# Create a MarkerCluster object
marker_cluster = MarkerCluster().add_to(site_map)

# 🟢🔴 Create a new column for marker color based on Class value
spacex_df['marker_color'] = spacex_df['class'].apply(lambda x: 'green' if x == 1 else 'red')

# 🚀 For each launch record, add a marker to the cluster
for index, record in spacex_df.iterrows():
    # Coordinates for the launch
    coordinate = [record['Lat'], record['Long']]
    
    # Marker popup text for details
    popup_text = f"""
    <b>Launch Site:</b> {record['Launch Site']}<br>
    <b>Outcome:</b> {'Success ✅' if record['class'] == 1 else 'Failure ❌'}
    """

    # Create and add marker to cluster
    marker = folium.Marker(
        location=coordinate,
        popup=folium.Popup(popup_text, max_width=300),
        icon=folium.Icon(color='white', icon_color=record['marker_color'], icon='rocket', prefix='fa')
    )
    
    marker_cluster.add_child(marker)

# Add the cluster layer to map
site_map.add_child(marker_cluster)

# Display the interactive map
site_map
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
from math import sin, cos, sqrt, atan2, radians
import folium
from folium.features import DivIcon

# --- Define the distance function (if not already defined) ---
def calculate_distance(lat1, lon1, lat2, lon2):
    R = 6373.0  # radius of Earth in km
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    distance = R * c
    return distance

# --- Define coordinates ---
launch_site_lat, launch_site_lon = 28.5623, -80.5774       # CCAFS SLC 40
coastline_lat, coastline_lon = 28.56367, -80.57163         # Found using MousePosition

# --- Calculate the distance ---
distance_coastline = calculate_distance(
    launch_site_lat, launch_site_lon,
    coastline_lat, coastline_lon
)
print(f"Distance from Launch Site to Coastline: {distance_coastline:.2f} km")

# --- Create Folium map centered at the launch site ---
distance_map = folium.Map(location=[launch_site_lat, launch_site_lon], zoom_start=13)
```

```python
from math import sin, cos, sqrt, atan2, radians
import folium
from folium.features import DivIcon

# --- Define the Haversine distance function ---
def calculate_distance(lat1, lon1, lat2, lon2):
    R = 6373.0  # radius of Earth in km
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    return R * c

# --- Launch site: CCAFS SLC 40 ---
launch_site_lat, launch_site_lon = 28.5623, -80.5774

# --- Coordinates of nearby features (approximate or from MousePosition) ---
coastline_lat, coastline_lon = 28.56367, -80.57163
city_lat, city_lon = 28.3922, -80.6077        # Cocoa Beach
railway_lat, railway_lon = 28.5721, -80.5853  # Nearby rail
highway_lat, highway_lon = 28.513, -80.655    # Nearby highway

# --- Compute all distances ---
distance_coastline = calculate_distance(launch_site_lat, launch_site_lon, coastline_lat, coastline_lon)
distance_city = calculate_distance(launch_site_lat, launch_site_lon, city_lat, city_lon)
distance_railway = calculate_distance(launch_site_lat, launch_site_lon, railway_lat, railway_lon)
distance_highway = calculate_distance(launch_site_lat, launch_site_lon, highway_lat, highway_lon)

# --- Create a Folium map centered at the launch site ---
distance_map = folium.Map(location=[launch_site_lat, launch_site_lon], zoom_start=12)

# Add launch site marker
folium.Marker(
    [launch_site_lat, launch_site_lon],
    popup='Launch Site: CCAFS SLC 40',
    icon=folium.Icon(color='blue', icon='rocket')
).add_to(distance_map)

# --- Define all nearby points and their distances ---
proximities = {
    "Coastline": ([coastline_lat, coastline_lon], distance_coastline, 'gray'),
    "City (Cocoa Beach)": ([city_lat, city_lon], distance_city, 'green'),
    "Railway": ([railway_lat, railway_lon], distance_railway, 'purple'),
    "Highway": ([highway_lat, highway_lon], distance_highway, 'red')
}

# --- Loop through each proximity point ---
for name, (coords, distance, color) in proximities.items():
    # Add marker
    folium.Marker(
        coords,
        popup=f"{name}: {distance:.2f} km",
        icon=folium.Icon(color=color, icon='info-sign')
    ).add_to(distance_map)
    
    # Add distance label
    folium.Marker(
        coords,
        icon=DivIcon(
            icon_size=(20,20),
            icon_anchor=(0,0),
            html=f'<div style="font-size: 12px; color:#d35400;"><b>{distance:.2f} KM</b></div>'
        )
    ).add_to(distance_map)
    
    # Draw line connecting the launch site to the proximity
    folium.PolyLine(
        locations=[[launch_site_lat, launch_site_lon], coords],
        color=color,
        weight=2.5,
        opacity=0.7
    ).add_to(distance_map)

# --- Display the interactive map ---
distance_map
```
